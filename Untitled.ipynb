```python
import pandas as pd
import numpy as np
import my_speller
```

```python
fn_dict = 'data/word_freq_dict.csv.zip'
fn_letters = 'data/letter_typos_matrix.json'
fn_typical = 'data/typical_typos.json'
fn_bigrams = 'data/bigrams.zip'
speller = my_speller.Speller(fn_dict, fn_letters, fn_typical, fn_bigrams, fix_register=True, verbose=False, try_split=True)
```

```python
df = pd.read_csv('sdf2.csv')
df.head()
```

```python
for mes in df.dropna().MESSAGE_CLIENT.values:
    if np.random.uniform(0, 1, 1)[0] < 0.05:
        res = speller(mes)
        if res != mes:
            print('Исходное:\n', mes, '\nИсправленное:\n', speller(mes))
            print('\n' + '-'*20 + '\n')
```
